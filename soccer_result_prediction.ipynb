# Hücum ve Defans Güçlerinin Hesaplanması

```python
df = pd.DataFrame(dataset)
table = pd.DataFrame(columns=('Team','HGS','AGS','HAS','AAS','HGC','AGC','HDS','ADS'))

#Veri setinde bulunan tüm maçlar ele alıanarak takımların evde ve deplasman attığı ve yediği gollerin ortalaması bulunur.

avg_home_scored = df.FTHG.sum() / 3465.0
avg_away_scored = df.FTAG.sum() / 3465.0
avg_home_conceded = avg_away_scored
avg_away_conceded = avg_home_scored

res_home = df.groupby('HomeTeam')
res_away = df.groupby('AwayTeam')

table.Team = res_home.mean().T.columns
table.HGS = res_home.FTHG.sum().values
table.HGC = res_home.FTAG.sum().values
table.AGS = res_away.FTAG.sum().values
table.AGC = res_away.FTHG.sum().values

"""
    HAS : Home Attack Strength, Ev Atak Gücü : Ev sahibi takımların attığı toplam gol 1 sezonda oynanan maç sayısına bölünerek 
    sezonluk ortalama bulunur. Bu ortalama da tüm maçlara göre bulunan ortalamaya bölünerek HAS bulunur.
    
"""
table.HAS = (table.HGS / 204.0) / avg_home_scored
table.AAS = (table.AGS / 204.0) / avg_away_scored
table.HDS = (table.HGC / 204.0) / avg_home_conceded
table.ADS = (table.AGC / 204.0) / avg_away_conceded

feature_table = df.iloc[:,:23]
feature_table = feature_table[['EvID','HomeTeam','AwayTeam','DepID','FTHG','FTAG','FTR','HST','AST','HC','AC']]
f_HAS = []
f_HDS = []
f_AAS = []
f_ADS = []
for index,row in feature_table.iterrows():
    f_HAS.append(table[table['Team'] == row['HomeTeam']]['HAS'].values[0])
    f_HDS.append(table[table['Team'] == row['HomeTeam']]['HDS'].values[0])
    f_AAS.append(table[table['Team'] == row['AwayTeam']]['AAS'].values[0])
    f_ADS.append(table[table['Team'] == row['AwayTeam']]['ADS'].values[0])

feature_table['HAS'] = f_HAS
feature_table['HDS'] = f_HDS
feature_table['AAS'] = f_AAS
feature_table['ADS'] = f_ADS

feature_table = feature_table[["HAS","HDS","AAS","ADS"]]
feature_table
```

```python
corner_model_data = pd.concat([dataset[["HomeTeam","AwayTeam","HC"]].assign(home=1).rename(
    columns={"HomeTeam" : "team", "AwayTeam" : "opponent","HC":"corners"}),
                             dataset[["AwayTeam","HomeTeam","AC"]].assign(home=0).rename(
                             columns={"AwayTeam":"team","HomeTeam" : "opponent","AC":"corners"})])

import statsmodels.api as sm
import statsmodels.formula.api as smf

poisson_model = smf.glm(formula = "corners ~ home + team + opponent",data = corner_model_data,
                       family = sm.families.Poisson()).fit()
```

```python
LRresult = (poisson_model.summary2().tables[1])
poisson_model.summary()
```

```python
coef = LRresult["Coef."].tolist()
intcep = coef[0]
homeadv  = coef[63]
median = LRresult["Coef."].median()

kornerdataset = dataset[['EvID','HomeTeam','AwayTeam','DepID','FTHG','FTAG','FTR','HTHG','HTAG','HTR','HS','AS','HST','AST','HF','AF','HC','AC','HY','AY','HR','AR']]
teamcoef = coef[1:32]
oppocoef = coef[32:63]
sirali_evcoef=[]
for i in kornerdataset["EvID"]:
    if i == "88":
        sirali_evcoef.append(median)
        continue
    elif i==31:
        sirali_evcoef.append(teamcoef[i-2])
    else:
        sirali_evcoef.append(teamcoef[i-1])
        
sirali_evoppocoef = []
for i in kornerdataset["EvID"]:
    if i == "88":
        sirali_evoppocoef.append(median)
        continue
    elif i==31:
        sirali_evoppocoef.append(oppocoef[i-2])
    else:
        sirali_evoppocoef.append(oppocoef[i-1])
        
sirali_depteamcoef = []
for i in kornerdataset["DepID"]:
    if i == "88":
        sirali_depteamcoef.append(median)
        continue
    elif i==31:
        sirali_depteamcoef.append(teamcoef[i-2])
    else:
        sirali_depteamcoef.append(teamcoef[i-1])

                
sirali_depoppocoef = []
for i in kornerdataset["DepID"]:
    if i == "88":
        sirali_depoppocoef.append(median)
        continue
    elif i==31:
        sirali_depoppocoef.append(oppocoef[i-2])
    else:
        sirali_depoppocoef.append(oppocoef[i-1])
        

kornerdataset.insert(1,"EvTeamCoef",sirali_evcoef,True)
kornerdataset.insert(2,"EvOppoCoef",sirali_evoppocoef,True)
kornerdataset.insert(6,"DepTeamCoef",sirali_depteamcoef,True)
kornerdataset.insert(7,"DepOffoCoef",sirali_depoppocoef,True)
#Hesaplamalar

EHC = []
EAC = []

import math
for i in range(0,3465):
    EHC.append(math.ceil(math.exp(intcep+homeadv+kornerdataset["EvTeamCoef"][i]+kornerdataset["DepOffoCoef"][i])))
    EAC.append(math.ceil(math.exp(intcep+kornerdataset["DepTeamCoef"][i]+kornerdataset["EvOppoCoef"][i])))
    
kornerdataset.insert(10,"EHC",EHC,True)
kornerdataset.insert(11,"EAC",EAC,True)
kornerdataset = kornerdataset[["EHC","EAC"]]

kornerdataset.insert(0,"HC",dataset["HC"],True)
kornerdataset.insert(3,"AC",dataset["AC"],True)

MAE = []
MSE = []
Fark = []
for i in range(len(kornerdataset)):
    MAE.append(abs(kornerdataset["HC"][i]-kornerdataset["EHC"][i]))
    MSE.append((kornerdataset["HC"][i]-kornerdataset["EHC"][i])**2)
    Fark.append(kornerdataset["EHC"][i] - kornerdataset["HC"][i])

kornerdataset.insert(4,"MAE",MAE,True)
kornerdataset.insert(5,"MSE",MSE,True)
kornerdataset.insert(6,"Fark",Fark,True)
```
